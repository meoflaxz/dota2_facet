```python
import polars as pl
df = pl.read_csv(r"heroes_data.csv")

df 
```

```python
df = df.with_columns(
    pl.col("Win Rate").str.replace_all(r"%$", ""),
    pl.col("No Matches").str.replace_all(r",", ""),
    pl.col("Hero Name").str.replace_all(r"-", " ").str.to_titlecase()
)

# df = df.with_columns(new_column.alias(column_name))

df
```

```python
rows_list = []

# Iterate over each row in the DataFrame
for idx in range(len(df)):
    # Get the row at the current index as a list and append it to rows_list
    rows_list.append(list(df[idx]))

# Print the list of rows
print(rows_list)
```

```python
df.write_csv("cleaned_heroes.csv")
```

```python
csv_file = 'cleaned_heroes.csv'
df.write_csv(csv_file)

# Read CSV file
csv_df = pl.read_csv(csv_file)

# Convert CSV DataFrame to SQL insert statements
rows = csv_df.select_at_idx(1).to_list()  # select the data as list of tuples
table_name = 'your_table_name'
sql_inserts = []
for row in rows:
    values = ', '.join(str(value) for value in row)
    sql_insert = f"INSERT INTO {table_name} VALUES ({values});"
    sql_inserts.append(sql_insert)

# Print or save SQL insert statements
for sql_insert in sql_inserts:
    print(sql_insert)
```
